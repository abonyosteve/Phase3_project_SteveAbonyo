# Data Preparation

```python
# This are some of the libraries we will be using for the project.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn import datasets
import statsmodels
from statsmodels.formula.api import ols
```

```python
# This functions calls the files to our colab environment
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_csv('Tel_Data.csv')
```

```python
# Fist we start by checking the percentage of missing values

total_rows = len(df)
missing_percentage = (df.isnull().sum() / total_rows) * 100

for column, percentage in missing_percentage.items():
    print(f"{column}, Percentage of Missing Values: {percentage:.2f}%")
```

## Data Processing

```python
from sklearn.preprocessing import LabelEncoder
# we used Label Encoding to convert the categorical variables
label_encoder = LabelEncoder()
df["international plan"] = label_encoder.fit_transform(df["international plan"])
df["voice mail plan"] = label_encoder.fit_transform(df["voice mail plan"])
```

```python
# We then convert the boolean values in 'churn' to 1s and 0s
df["churn"] = df["churn"].astype(int)
```

```python
from sklearn.model_selection import train_test_split

# We then drop unnecessary columns and split the data
X = df.drop(["churn", "state", "phone number"], axis=1)
y = df["churn"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
df.head(3)
```

## Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Model 1: Logistic Regression (Simple Model)
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred_logistic = logistic_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_logistic)
classification_report_str = classification_report(y_test, y_pred_logistic)

# Print the model performance metrics
print("Logistic Regression Model Performance:")
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", classification_report_str)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Model 2: Random Forest Classifier
random_forest_model = RandomForestClassifier(random_state=42)
random_forest_model.fit(X_train, y_train)
y_pred_random_forest = random_forest_model.predict(X_test)

# Evaluate the Random Forest model
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)
classification_report_random_forest = classification_report(y_test, y_pred_random_forest)

# Print the Random Forest model performance metrics
print("Random Forest Model Performance:")
print(f"Accuracy: {accuracy_random_forest:.2f}")
print("Classification Report:\n", classification_report_random_forest)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Model 3: hyperparameters
# First we define hyperparameters to tune
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [10, 20, 30],       # Maximum depth of each tree
    'min_samples_split': [2, 5, 10]  # Minimum number of samples required to split an internal node
}

# Then we create a Random Forest classifier
random_forest_model_tuned = RandomForestClassifier(random_state=42)

# Then perform Grid Search to find the best hyperparameters
grid_search = GridSearchCV(random_forest_model_tuned, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# This is followed by getting the best parameters from the grid search
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Then we fit the Random Forest model with the best hyperparameters
best_random_forest_model = RandomForestClassifier(random_state=42, **best_params)
best_random_forest_model.fit(X_train, y_train)
y_pred_best_random_forest = best_random_forest_model.predict(X_test)

# We finally Evaluate the tuned Random Forest model
accuracy_best_random_forest = accuracy_score(y_test, y_pred_best_random_forest)
classification_report_best_random_forest = classification_report(y_test, y_pred_best_random_forest)

# Print the tuned Random Forest model performance metrics
print("Tuned Random Forest Model Performance:")
print(f"Accuracy: {accuracy_best_random_forest:.2f}")
print("Classification Report:\n", classification_report_best_random_forest)
```
