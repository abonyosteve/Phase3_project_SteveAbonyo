##### Importing the files I saved in RDS format

```python
df_crashes = feather.read_feather('df_crashes_sampledata.rds')
df_vehicles = feather.read_feather('df_vehicles_sampledata.rds')
df_people = feather.read_feather('df_people_sampledata.rds')
print(df_crashes.shape)
print(df_vehicles.shape)
print(df_people.shape)
```

#### ***Merging Data***

```python
##Step 1
df1 = pd.merge(left=df_crashes, right = df_vehicles, left_on='CRASH_RECORD_ID', right_on="CRASH_RECORD_ID")
#Step 2
df = pd.merge(left=df1, right=df_people, left_on = 'VEHICLE_ID', right_on='VEHICLE_ID')
#Step 3
print(df.shape)
#Step 4
df.head()
```

#### ***Checking for Null Values***

```python
null_values = df.isna().sum()
null_perct = (null_values[null_values>0] / len(df)) * 100
null_perct.to_frame('Percentage of null values').style.background_gradient(cmap='Blues')
```

#### ***Dropping colmns with 50% plus Null Values***

```python
cols_to_remove = null_perct[null_perct>=50].index.tolist()
df = df.drop(columns = cols_to_remove)
print(df.shape)
```
